```python
from maze import Maze
from algo import a_star, dijkstra, bfs
```

```python
maze_breadth = 5
maze_length = 6

start_pos = (0, 0)
goal_pos = (4, 5)

obstacles = [
    (0, 1),
    (2, 1),
    (2, 3),
    (3, 1),
    (3, 4),
    (4, 4)
]

# Create a maze instance
maze = Maze(maze_breadth, maze_length)

maze.set_start(*start_pos)
maze.set_goal(*goal_pos)

# Set obstacles in the maze
for obstacle_pos in obstacles:
    maze.set_obstacle(*obstacle_pos)

# Print the maze
print(maze)
```

```python
%%time
bfs_path = bfs(maze)

print("BFS Path:", bfs_path)
```

```python
bfs(maze, visualize=True)
```

```python
%%time
dijkstra_path = dijkstra(maze)

print("A* Path:", dijkstra_path)
```

```python
maze.clear_visited()
dijkstra(maze, visualize=True)
```

```python
%%time
a_star_path = a_star(maze)

print("A* Path:", a_star_path)
```

```python
maze.clear_visited()
a_star(maze, visualize=True)
```
